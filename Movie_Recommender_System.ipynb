```python
# Importing the required libraries

import numpy as np
import pandas as pd
```

```python
# Loading the dataset
credits = pd.read_csv('/content/drive/MyDrive/tmdb_5000_credits.csv')
movies = pd.read_csv('/content/drive/MyDrive/tmdb_5000_movies.csv')
```

```python
# Displaying first 5 rows of movies dataset
movies.head()
```

```python
# Displaying first 5 rows of credits dataset

credits.head()
```

```python
# Merging the 2 datasets

movies = movies.merge(credits, on='title')
```

```python
# To shortlist columns in dataset

movies.info()
```

```python
# Reason for not keeping original_language column
# Data is highly biased towards English
movies['original_language'].value_counts()
```

```python
# Filtering only the required columns for the required columns

# Required columns : genres, id, keywords, title, overview, cast, crew

movies = movies[['movie_id', 'title', 'overview','genres', 'keywords', 'cast', 'crew']]
```

```python
# Final dataframe after filtering out columnns

movies.head()
```

```python
# To check for missing data

movies.isnull().sum()
```

```python
# To drop the 3 columns for which overiew is not available
movies.dropna(inplace=True)
```

```python
# To check for duplicates

movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# Every entry in genre is in string data type so ast is imported to convert it into list
import ast

```

```python
'''[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"},
 {"id": 878, "name": "Science Fiction"}]'''
# Convert above into ['Action', 'Adventure', 'Fantasy', 'Science Fiction']

# Helper function for required conversion
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L



```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# Helper function to get cast in required format
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter += 1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# Helper function to get the name of the director for every movie
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# Convert every entry of overview from string into a list

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", '') for i in x])
```

```python
# Concatenate columns into a tag column

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

Text Vectorization -- Technique used -> Bag of Words

```python
# To get different word forms in their root form
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Helper function to get words in root form
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
# CounterVectorizer creates a pool of words from every entry in the tags column
# 5000 most occuring words are obtained
# 5000 dimension vector is created for each movie for each one of the 5000 most occuring word

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
# To compute similarity using the 5000 dimensional vectors created
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# Function to recommend movie

def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
